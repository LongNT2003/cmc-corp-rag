```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import time
import os
```

```python
def crawl_introduction(url):
    options = Options()     
    options.add_experimental_option('detach', True)    
    driver = webdriver.Edge(options=options) 
    driver.maximize_window()
    driver.get(url)
    content=[]
    about = driver.find_elements(By.XPATH, '//*[@id="content"]/div[1]/div/div[2]/div')
    about=[each.text.replace('\n',' ') for each in about]
    content.extend(about)

    about2=driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div/div/div')
    content.append(about2.text.replace('\n',' '))

    about3=driver.find_element(By.XPATH,'//*[@id="content"]/div[3]/div/div/div[2]')
    content.append(about3.text)

    about4_title=driver.find_element(By.XPATH,'//*[@id="content"]/div[4]/div/div/h3').text
    about4_content=driver.find_elements(By.XPATH,'//*[@id="content"]/div[4]/div/div/div/div/div')
    about4=[about4_title+':'+each.text for each in about4_content]
    content.extend(about4)

    top_down=driver.find_element(By.XPATH,'//*[@id="content"]/div[5]/div/div[1]/h2/em').click()
    bo_phan=driver.find_elements(By.XPATH,'//*[@id="content"]/div[5]/div/div[1]/ul/li')
    for board_id,bo in enumerate(bo_phan):
        time.sleep(0.5)
        bo.click()
        time.sleep(0.5)
        leaders=driver.find_elements(By.XPATH,f'//*[@id="board{board_id+1}"]/div/div[2]/ul/li/a')
        for idx,leader in enumerate(leaders):
            actions = ActionChains(driver)
            actions.move_to_element(leader).perform()
            leader.click()
            time.sleep(0.5)
            leader_career=driver.find_element(By.XPATH,f'//*[@id="board_1_{idx+1}"]')
            print(leader_career.text)
            content.append(leader_career.text.replace('\n',','))
    driver.close()
    return '\n'.join(content)

def crawl_history(url):
    options = Options()     
    options.add_experimental_option('detach', True)    
    driver = webdriver.Edge(options=options) 
    driver.maximize_window()
    driver.get(url)
    intro=driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div/div/div[2]')
    total=[intro.text]
    num_of_years=len(driver.find_elements(By.XPATH,'//*[@id="content"]/div[2]/div/div/div/div'))
    for i in range(num_of_years):
        try:
            right_move=driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div/button[2]')
            right_move.click()
            time.sleep(1)
            year_descriptions=driver.find_element(By.XPATH,f'//*[@id="content"]/div[2]/div/div/div/div[{i+1}]')
            total.append(year_descriptions.text)
        except:
            pass
    driver.close()
    return '\n'.join(total)

def crawl_achivement(url):
    options = Options()     
    options.add_experimental_option('detach', True)    
    driver = webdriver.Edge(options=options) 
    driver.maximize_window()
    driver.get(url)
    intro=driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div/div/div[2]')

    total=[intro.text]
    menu = driver.find_element(By.CLASS_NAME, 'menu__pc')
    driver.execute_script("arguments[0].scrollIntoView(true);", menu)
    num_of_years=len(driver.find_elements(By.XPATH,'//*[@id="content"]/div[2]/div[2]/div/div/div/div'))
    for i in range(num_of_years):
        try:
            right_move=driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div[2]/div/button[2]')
            right_move.click()
            time.sleep(1)
            year_achive=driver.find_element(By.XPATH,f'//*[@id="content"]/div[2]/div[2]/div/div/div/div[{i+1}]')
            years=driver.find_element(By.CLASS_NAME,'content__award__main__item__year')
            print(years)
            total.append(year_achive.text)
        except:
            pass
    driver.close()
    return '\n'.join(total)
def about_cmc_contents():
    urls = [
        'https://www.cmc.com.vn/introduction',
        'https://www.cmc.com.vn/history',
        'https://www.cmc.com.vn/achievement'
    ]
    
    introduction = crawl_introduction(urls[0])
    history = crawl_history(urls[1])
    achievement = crawl_achivement(urls[2])
    
    combined_content = [introduction,history,achievement]
    return combined_content
about_cmc=about_cmc_contents()
```

```python
def insight_crawl(url):
    options = Options()     
    options.add_experimental_option('detach', True)    
    prefs = {"profile.managed_default_content_settings.images": 2}
    options.add_experimental_option("prefs", prefs)
    driver = webdriver.Edge(options=options) 
    driver.maximize_window()
    driver.get(url)
    popular_content=driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div/div/div[2]/div').text
    print(popular_content)
    direct=driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div/div/div[1]/div/div[1]')
    direct.click()
    news=[popular_content]
    while 1:
        try:
            time.sleep(1)
            content=driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div/div/div[1]')
            news.append(content.text)
            next_page=driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div/div/div[1]/div/div/div[1]/a')
            next_page.click()
        except:
            break
    driver.quit()
    return news
def remove_consecutive_duplicates(text):
    # Chia đoạn văn thành từng câu
    lines = text.split('\n')
    
    # Tạo một danh sách mới để lưu các câu không bị trùng lặp
    filtered_lines = []
    
    for i in range(len(lines)):
        # Nếu câu này khác câu trước đó, hoặc nếu nó là câu đầu tiên, thêm nó vào danh sách mới
        if i == 0 or lines[i] != lines[i - 1]:
            filtered_lines.append(lines[i])
    
    # Gộp lại các câu đã lọc và trả về
    return '\n'.join(filtered_lines)


```

```python
thong_cao=insight_crawl('https://www.cmc.com.vn/insight/thong-cao-bao-chi')
```

```python
su_kien=insight_crawl('https://www.cmc.com.vn/insight/su-kien')
```

```python
def insight_contents(d1,d2,d3):
    urls = [
        d1,
        d2,
        d3
    ]
    
    all_content = []
    
    for data in urls:
        data=[remove_consecutive_duplicates(each).replace('SHARE','') for each in data]
        all_content.extend(data)
    
    return all_content
insight=insight_contents(tin_tuc,thong_cao,su_kien)
```

```python
def solution_and_service_crawl(url):
    options = Options()     
    options.add_experimental_option('detach', True)    
    driver = webdriver.Edge(options=options) 
    driver.maximize_window()
    driver.get(url)
    description=driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div')
    detail=[description.text]
    num_sols=len(driver.find_elements(By.XPATH,'//*[@id="content"]/div[2]/div/div/div/div'))
    try:
        for sol in range(num_sols):
            time.sleep(1)
            sol_detail=driver.find_element(By.XPATH,f'//*[@id="content"]/div[2]/div/div/div/div[{sol+1}]')
            detail.append(sol_detail.text)
            driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div/button[2]').click()
    except:
        sol_detail=driver.find_elements(By.XPATH,'//*[@id="content"]/div[2]/div')
        detail+=[text.text for text in sol_detail]
    driver.close()
    print('.'.join(detail))
    return  '.'.join(detail)

def service_and_product_contents():
    urls = [
        'https://www.cmc.com.vn/technology-and-solution',
        'https://www.cmc.com.vn/telecommunication',
        'https://www.cmc.com.vn/global-business',
        'https://www.cmc.com.vn/cmc-technology-institute'
    ]
    
    all_solutions = []
    
    for url in urls:
        options = Options()     
        options.add_experimental_option('detach', True)    
        driver = webdriver.Edge(options=options) 
        driver.maximize_window()
        driver.get(url)
        
        num_of_ser = len(driver.find_elements(By.XPATH, '//*[@id="content"]/div[2]/div[3]/div/div/div[2]/div/div[1]/div/div[1]/div/div'))
        thong_tin_giai_phap = []
        
        for ser in range(num_of_ser):
            try:
                href = driver.find_element(By.XPATH, f'//*[@id="content"]/div[2]/div[3]/div/div/div[2]/div/div[1]/div/div[1]/div/div[{ser+1}]/div/div/h3/a').get_attribute('href')
                print(href)
            except:
                pass
                continue
            if href is not None:
                thong_tin_giai_phap.append(solution_and_service_crawl(href))
        
        driver.quit()
        print(thong_tin_giai_phap)
        # Merge the results from the current URL into the overall list
        all_solutions.extend(thong_tin_giai_phap)
    
    # Combine all solutions from all URLs into a single string
    return all_solutions

# Execute the function
ser_prod = service_and_product_contents()
```

```python
def solution_and_story_crawl(url):
    options = Options()     
    options.add_experimental_option('detach', True)    
    driver = webdriver.Edge(options=options) 
    driver.maximize_window()
    driver.get(url)
    description=driver.find_element(By.XPATH,'//*[@id="content__ajax"]/div[1]/div/div[1]/div/div[2]')
    result=[description.text.replace('\n','.')]
    num_solutions=len(driver.find_elements(By.XPATH,'//*[@id="content__ajax"]/div[1]/div/div[2]/div/div/div/div'))
    for idx in range(num_solutions):
        if idx>=4:
            driver.find_element(By.XPATH,'//*[@id="content__ajax"]/div[1]/div/div[2]/div/button[2]').click()
            time.sleep(0.5)

        curr=driver.find_element(By.XPATH,f'//*[@id="content__ajax"]/div[1]/div/div[2]/div/div/div/div[{idx+1}]')
        result.append(curr.text.replace('\n',','))

    num_stories=len(driver.find_elements(By.XPATH,'//*[@id="content__ajax"]/div[2]/div/div/div/div'))
    for story in range(num_stories):
        try:
            story_link=driver.find_element(By.XPATH,f'//*[@id="content__ajax"]/div[2]/div/div/div/div[{story+1}]/div/div/div/div/a[2]').get_attribute('href')
            print(story_link)
        except:
            story_link=driver.find_element(By.XPATH,f'//*[@id="content__ajax"]/div[2]/div/div/div/div[{story+1}]/div/div/div/div/div/a').get_attribute('href')
            print(story_link)
        
        try:
            result.append(case_detail(story_link))
        except:
            pass

        if story <num_stories-1:
            next_story=driver.find_element(By.XPATH,'//*[@id="content__ajax"]/div[2]/div/button[2]')
            next_story.click()
            time.sleep(0.5)
    
    driver.quit()
    return '\n'.join(result)

def case_detail(url):
    options = Options()     
    options.add_experimental_option('detach', True)    
    driver = webdriver.Edge(options=options) 
    driver.maximize_window()
    driver.get(url)
    title=driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div/div/div[1]/div[1]')
    result=[title.text.replace('\n','.')]
    
    challenge=driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div/div/div[1]/div[2]')
    result.append(challenge.text.replace('\n','.'))

    solution=driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div/div/div[1]/div[3]')
    solution=solution.text.replace('\n','.')

    features=driver.find_elements(By.XPATH,'//*[@id="content"]/div[1]/div/div/div[2]/div/div/p')
    features='.'.join([feature.text for feature in features])
    result.append(solution+'\n'+features)
    
    driver.close()
    return '\n'.join(result)

def solution_and_story_content():
    urls=[
        'https://www.cmc.com.vn/cross-industry',
        'https://www.cmc.com.vn/government',
        'https://www.cmc.com.vn/public-sector',
        'https://www.cmc.com.vn/manufacturing',
        'https://www.cmc.com.vn/banking',
        'https://www.cmc.com.vn/insurance',
        'https://www.cmc.com.vn/financial-services',
        'https://www.cmc.com.vn/education'
        
    ]
    all_contents=[]
    for link in urls:
        all_contents.append(solution_and_story_crawl(link))
    return all_contents

sol_story=solution_and_story_content()
print(sol_story[0])
```
